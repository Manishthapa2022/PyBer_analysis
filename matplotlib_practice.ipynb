```python
# import dependecies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create a plot and add a label for the legend, adding markers, grid
plt.plot(x_axis, y_axis, label="Boston", color="red", linewidth=5, marker="*")
plt.xlabel("Months of year")
plt.ylabel("Fare($)")
plt.title("Fares over months")
# creating the limits for x and y axis
plt.ylim(0, 45)
plt.xlim("Jan", "June")
# change the figure size in pixels
# plt.figure(figsize=(20,10))

#adding a grid to the graph
plt.grid()

# Adding legend to the graph
plt.legend()

```

```python
# creating Bar charts using MATLAB method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create a plot
plt.barh(x_axis, y_axis,color="Green", label="Boston")
# to get the correct order of months in the axes
plt.gca().invert_yaxis()
plt.ylabel("Months")
plt.xlabel("Fares ($)")
plt.title("Fares against Months")

plt.legend()
```

```python
# Using scatter plot to visivulize data

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Increase the data size in y axis by 3 times so that the graph is more legibl
plt.scatter(x_axis, y_axis, color="r", label="Chicago",edgecolors="black", s=[i*3 for i in y_axis], marker="o")
plt.xlabel("Months in year")
plt.ylabel("Fares ($)")
plt.title('Fares against the months')

plt.legend()
```

```python
# creating pie charts
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct="%1.1f%%", colors=colors)


plt.show()
```
